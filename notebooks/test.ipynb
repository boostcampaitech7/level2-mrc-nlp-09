```python
csv_file = "../data/preprocessed/densed_BM25Ensemble_topk_100.csv"
dataset = load_from_disk("../data/raw/train_dataset/") 
validation_df = pd.DataFrame(dataset['validation'])
df = pd.read_csv(csv_file)


df_merged = pd.merge(df, validation_df[['id', 'answers']], on='id', how='left')

# context1, context2 ... 
context_columns = [col for col in df_merged.columns if col.startswith('top')][:10]  # Identify all context columns [:n]으로 상위 n까지만 획득 가능
df_merged['context'] = df_merged[context_columns].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)

```

```python
# 'answers'에서 'text' 추출
df_merged['answer_text'] = df_merged['answers'].apply(lambda x: x['text'][0] if isinstance(x, dict) else '')

# 각 context에 answer_text가 포함되어 있는지 여부 확인
df_merged['answer_in_context'] = df_merged.apply(lambda row: row['answer_text'] in row['context'], axis=1)

# answer_in_context 열을 출력하여 확인
print(df_merged[['id', 'answer_text', 'context', 'answer_in_context']].head())
```

```python

# 포함된 경우와 포함되지 않은 경우의 비율 계산
answer_in_context_counts = df_merged['answer_in_context'].value_counts(normalize=True)

# 비율 출력
print(f"답변이 포함된 context 비율: {answer_in_context_counts[True]:.2%}")
print(f"답변이 포함되지 않은 context 비율: {answer_in_context_counts[False]:.2%}")
```
